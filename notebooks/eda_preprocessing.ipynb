```python
import pandas as pd

df = pd.read_csv("data/raw/complaints.csv")
df.head()
df.info()
```

```python
df['product'].value_counts()
```

```python
df['narrative_length'] = df['consumer_complaint_narrative'].astype(str).apply(lambda x: len(x.split()))
df['narrative_length'].describe()
```

```python
df['consumer_complaint_narrative'].isna().sum()
```

```python
products = [
    "Credit card",
    "Personal loan",
    "Savings account",
    "Money transfer"
]

df = df[df['product'].isin(products)]
df = df.dropna(subset=['consumer_complaint_narrative'])
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'[^a-z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_narrative'] = df['consumer_complaint_narrative'].apply(clean_text)
```

```python
df.to_csv("data/processed/filtered_complaints.csv", index=False)
```
